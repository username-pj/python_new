**Total score       Grading**

- above 250 = A Grade

- 200 to 250 = B Grade

- 150 to 200 = C Grade

- 100 to 150 = D Grade

- 50 to 100 = E Grade

- below 50 = F Grade ( means Fail )

```python
def GetGrade(Totalscore):
    if ( Totalscore >= 250 ):
        return 'A'
    elif ( Totalscore >=200):
        return 'B'
    elif ( Totalscore >= 150):
        return 'C'
    elif ( Totalscore >= 100):
        return 'D'
    elif (Totalscore >= 50):
        return 'E'
    else:
        return 'F'
```

```python
df1['Grade'] = df1['total score'].apply(lambda x : GetGrade(x))
```

```python
sns.countplot(x="Grade", data = df1, order=['A','B','C','D','E','F'],  palette="muted")
plt.show()
```

```python
plt.figure(figsize=(9,6))
sns.countplot(x='parental level of education', data = df1, hue='Grade',palette='Set2' )
plt.xticks(rotation=45)
```
